```python
from immortality import AdvancedDNAOptimizer
```

```python
# Ищем путь к бессмертию
# Создаем экземпляр оптимизатора с расширенными настройками
optimizer = AdvancedDNAOptimizer(
    population_size=20,          # Увеличенный размер популяции
    num_islands=4,               # 4 острова для разных стратегий эволюции
    mutation_rate_base=0.1,      # Базовая скорость мутации
    crossover_rate=0.7,          # Высокая вероятность скрещивания для разнообразия
    tournament_size=10,           # Размер турнира для отбора
    elitism_rate=0.1,            # Сохранение 10% лучших особей
    max_generations=50,          # Увеличенное количество поколений
    stagnation_limit=10,         # Лимит поколений без улучшения
    simulation_steps=500,        # Увеличенное количество шагов симуляции
    verbose=True
)

# Запускаем оптимизацию с расширенными диапазонами параметров
structure_params_range = {
    'penta_ratio': (0.2, 0.8),
    'hexa_ratio': (0.2, 0.8),
    'growth_factor': (0.3, 0.9),
    'regeneration_factor': (0.3, 0.9),
    'resonance_factor': (0.3, 0.9),
    'fractal_complexity': (0.3, 0.9),
    'energy_efficiency': (0.3, 0.9),
    'adaptability': (0.3, 0.9)
}

best_solution = optimizer.optimize(structure_params_range)

# Визуализируем прогресс оптимизации
optimizer.visualize_optimization_progress()

# Проводим подробную симуляцию лучшего организма
simulation_results = optimizer.simulate_best_solution(detailed=True, save_animation=True)
```
